```python
import pandas as pd
```

```python
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('/Users/alehmann/Documents/Documents/HIH/BDA/BDA2017/data/kmeans/2015-05-26_17-11-46.csv')
data.index = data['loggingTime']
data
```

```python
type(data)
```

```python
acc = data[['accelerometerAccelerationX','accelerometerAccelerationY','accelerometerAccelerationZ']]
```

```python
model = KMeans(n_clusters=6)
```

```python
model.fit(acc)
```

```python
model.cluster_centers_
```

```python
model.cluster_centers_
```

```python
res = model.predict(acc)
```

```python
res
```

```python
cluster_sizes = [len(res[res==x]) for x in range(6)]
```

```python
cluster_sizes
```

```python
length_centers = [norm(x) for x in model.cluster_centers_]
length_centers
```

```python
data['cluster'] = res
```

```python
cluster_speed = data[['locationSpeed','cluster']]
cluster_speed
```

```python
cluster_speed.plot(y='locationSpeed')
```

```python
for i in range(6):
    cluster_speed[f'cluster{i}'] = cluster_speed['locationSpeed'][cluster_speed['cluster']==i]
    
cluster_speed
```
